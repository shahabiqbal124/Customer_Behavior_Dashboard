```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r"[^\w]+", "_", regex=True)
      .str.replace(r"_+", "_", regex=True)
      .str.strip("_")
)
```

```python
df.columns
```

```python
# creating new columns to handle age group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# column for purchase frequencey days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

# --- Step 0: Setup your connection ---
username = "postgres"
password = "12345"
host = "localhost"
port = "5432"
database = "customer_behaviour"

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# --- Step 1: Optional: ensure public schema exists ---
with engine.connect() as conn:
    conn.execute(text("CREATE SCHEMA IF NOT EXISTS public;"))
    conn.commit()

# --- Step 2: Load DataFrame into PostgreSQL ---
# Your DataFrame 'df' should have 19 columns matching 'customer'
df.to_sql(
    name="customer",        # lowercase table name
    con=engine,
    if_exists="replace",    # overwrite if exists
    index=False
)

print("Data successfully loaded into table 'customer' in schema 'public'.")

# --- Step 3: Query back the table to verify ---
df_check = pd.read_sql("SELECT * FROM public.customer LIMIT 5", engine)
print(df_check.head())
```
